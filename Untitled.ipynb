```python
import matplotlib.pyplot as py
```

```python
x1=[2,3,4]
y1=[3,4,5]
x2=[2,2.5,3]
y2=[5,4,3]
x3=[2,3,4]
y3=[4,4,4]

py.xlabel('x-axis')
py.ylabel('y-axis')
py.plot(x1,y1,label='solomon', color='y')
py.plot(x2,y2,label='Joel', color='r')
py.plot(x3,y3,label='Richard', color='b')
py.grid(c='grey')
py.legend(loc='best')
py.show()
```

```python
x4=['solomon','Joel','Richard','Divine']
y4=[10,20,50,40]

py.xlabel('x-axis')
py.ylabel('y-axis')
py.bar(x4,y4,color='g')
py.grid(c='grey')
py.show()
```

```python
x1=[2,3,4,5,3,4]
y1=[3,4,5,2,3,4]

py.xlabel('x-axis')
py.ylabel('y-axis')
py.scatter(x1,y1,label='mines',color='red',marker='x')
py.legend()
py.grid(c='grey')
py.show()
```

```python

h=['Solomon','Divine','Joel','Rejoice','Richard']
i=[440,430,445,400,350]
colors=['red','magenta','cyan','yellow','green']
explode=(0.2,0,0,0,0)

py.pie(i,labels=h,explode=explode,colors=colors,shadow=True,autopct='%1.2f%%',startangle=90)
py.show()
```

```python
h=['Solomon','Divine','Joel','Rejoice','Richard']
i=[440,430,445,400,350]
colors=['red','magenta','cyan','yellow','green']

piece,text=py.pie(i,colors=colors,shadow=True,startangle=140)
py.legend(piece,h,loc='upper right')
py.show()
```
