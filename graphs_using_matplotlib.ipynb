```python
style.use('fivethirtyeight')
fig=py.figure()
ax1=fig.add_subplot(1,1,1)
def animations(i):
    graph_data=open('hello.txt','r').read()
    lines=graph_data.split('\n')
    xs=[]
    ys=[]
    for  line in lines:
        if len(line) > 1:
            x,y=line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs,ys)
    
ani=an.FuncAnimation(fig,animations,interval=1000)
py.show()
```
